```python
import numpy as np # linear algebra
import pandas as pd
```

```python
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
```

```python
dataset = pd.read_csv('train.csv')
X_train = dataset.iloc[:,[2,4,5,6,7,9,11]].values
y_train = dataset.iloc[:,1].values
testDataset=pd.read_csv('test.csv')
X_test=testDataset.iloc[:,[1,3,4,5,6,8,10]].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train[:,1] = le.fit_transform(X_train[:,1])
X_test[:,1] = le.fit_transform(X_test[:,1])
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_test[:, 2:3] = imputer.fit_transform(X_test[:, 2:3])
X_train[:, 2:3] = imputer.fit_transform(X_train[:, 2:3])
X_test[:, 5:6] = imputer.fit_transform(X_test[:,5:6])
imputer1 = SimpleImputer(strategy='most_frequent')
X_train[:, -1] = imputer1.fit_transform(X_train[:, -1].reshape(-1, 1)).ravel()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
X_train = np.array(ct.fit_transform(X_train))
X_test = np.array(ct.fit_transform(X_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators =10 ,max_depth=5, random_state = 1)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred = np.where(y_pred < 0.5, 0, 1)
print (y_pred)
```

```python
output = pd.DataFrame({'PassengerId': testDataset.PassengerId, 'Survived': y_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
